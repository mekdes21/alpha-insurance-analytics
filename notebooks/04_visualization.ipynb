```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('../data/insurance_data_cleaned.csv')

# Set figure size
plt.figure(figsize=(10, 6))  # Increase figure size

# Create boxplot
sns.boxplot(x='CoverCategory', y='TotalClaims', data=df)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')  # Rotate labels by 45 degrees, and adjust horizontal alignment

# Add title
plt.title('Claims by Cover Category')

# Save the plot as an image file
plt.savefig('../reports/plots/boxplot_claims.png')

# Show the plot
plt.tight_layout()  # Adjust the layout to ensure the labels fit within the figure
plt.show()
```
